# HW01 : Chatbot - Pizza Shop

```python
class Product:
    def __init__(self, id, name, desc, price):
        self.id = id
        self.name = name
        self.description = desc
        self.price = price
```

```python
class Order:
    def __init__(self, product: Product, amount):
        self.product = product
        self.amount = amount

    def cal_total(self):
        return self.product.price * self.amount
```

```python
class PizzaShop:
    divider = "---------------------"

    def __init__(self, shop_name):
        self.shop_name = shop_name
        self.menus = []


    def add_menu(self, id, name, desc, price: int):
        prod = Product(id, name, desc, price)
        self.menus.append(prod)

    def begin_order(self):
        print(f"Hi! Welcome to {self.shop_name}")
        self.print_menu()
        self.print_instruction()

        orders = []
        is_exit = False

        while not is_exit:
            order_id = input("Order: ")
            order_id = order_id.replace("[","").replace("]","").lower()
            if not order_id.isdigit():
                if order_id == "exit" or order_id == "e":
                    is_exit = True
                elif order_id == "m" or order_id == "menu":
                    self.print_menu()
                else:
                    self.print_instruction()
                continue

            prod = self.search_product(int(order_id))
            if prod == None:
                self.print_instruction()
                continue

            is_correct = False
            while not is_correct:
                print(f"You order for: {prod.name}")
                print("""How many you want to order?
    [h] for menu description
    [p] for menu price
    [c] for cancel and choose new menu""")
                amount = input("")
                if not amount.isdigit():
                    if amount == "h":
                        print(f"{prod.name} is {prod.description}")
                    elif amount == "p":
                        print(f"{prod.name} is {prod.price}")
                    elif amount == "c":
                        is_correct = True
                    else:
                        print("Please try again.")
                else:
                    o = Order(prod, int(amount))
                    orders.append(o)
                    is_correct = True

        self.sum_and_pay(orders)

        return

    def sum_and_pay(self, orders):
        print("Your order:")
        for o in orders:
            print(f"{o.product.name} | {o.amount} | {o.cal_total()}")
        print(f"Total : {self.cal_total(orders)}")

    def search_product(self, id):
        for m in self.menus:
            if id == m.id:
                return m
        return None

    def cal_total(self, orders):
        total = 0
        for o in orders:
            total += o.cal_total()
        return total

    def print_instruction(self):
        print("""If you want to order, please type
    [number] of menu
    [m] for menu
    [e]xit for check out.""")

    def print_menu(self):
        print("Our menus:")
        for m in self.menus:
            print(f"[{m.id}] {m.name} - {m.price}")
        print(self.divider)
```

```python
abc_pizza = PizzaShop("ABC Pizza Shop")
abc_pizza.add_menu(1, "Margherita", "Tomato sauce, Fiord latte' cheese, Parmesan cheese and Basi", 295)
abc_pizza.add_menu(2, "Bolognese", "Bolognese topping, Mozzallera cheese, Parmesan cheese and Tomato sauce", 380)
abc_pizza.add_menu(3, "Hawaiian", "Smoked Bacon, Ham, Pineapple, Mozarella cheese, Parmesan cheese and Tomato sauce", 350)
abc_pizza.add_menu(4, "Parma Ham & Rocket", "Parma Ham, Rocket and Tomato sauce", 460)
abc_pizza.add_menu(5, "Coke", "Coke can 330ml", 20)
```

```python
abc_pizza.begin_order()
```

# HW02 : Pao Ying Chup Game

```python
import random

class Game:
    def __init__(self):
        self.game_name: str = "Pao Ying Chup!"
        self.instruction: str = """You have to choose one of this choice: [R]ock , [S]cissor, [P]aper
    [R]ock win against [S]cissor
    [S]cissor win againt [P]aper
    [P]aper win against [R]ock

    Who win will get 1 score.

    If you want to read instruction again, please type [i].
    If you want to stop and view score summary, please type [exit]."""
        self.greeting: str = f"Welcome to {self.game_name}"
        self.play_choice = ["r", "p", "s"]
        self.txt_draw = "It seems we draw"
        self.txt_gm_win = "I win!"
        self.txt_player_win = "You win!"
        self.divider = "------------------------------------"

        self.gm_score = 0
        self.player_score = 0


    def start(self):
        print(self.greeting)

        self.player_name = input("What's your name? ")

        print(f"OK, {self.player_name} ", self.instruction)

        is_exit = False

        while not is_exit:
            gm_choice = random.choice(self.play_choice)

            player_choice = self.clean_choice(input(f"{self.player_name}, which one will you choose? "))

            if player_choice in self.play_choice:
                self.check(gm_choice, player_choice)
            elif player_choice == "i":
                print(self.instruction)
            elif player_choice == "exit" or player_choice == "e" or player_choice == "bye":
                is_exit = True
            else:
                print(self.instruction)

            self.print_divider()


        self.stop()


    def stop(self):
        print(f"{self.player_name} score is {self.player_score}")
        print(f"My score is {self.gm_score}")

        if self.gm_score == self.player_score:
            print(self.txt_draw)
        elif self.gm_score > self.player_score:
            print(self.txt_gm_win)
        else:
            print(self.txt_player_win)


    def check(self, gm_choice, player_choice):
        print(f"I choose {gm_choice}")
        if player_choice == "r":
            if gm_choice == player_choice:
                print(self.txt_draw)
            elif gm_choice == "p":
                self.gm_score += 1
                print(self.txt_gm_win)
            elif gm_choice == "s":
                self.player_score += 1
                print(self.txt_player_win)
        elif player_choice == "p":
            if gm_choice == player_choice:
                print(self.txt_draw)
            elif gm_choice == "s":
                self.gm_score += 1
                print(self.txt_gm_win)
            elif gm_choice == "r":
                self.player_score += 1
                print(self.txt_player_win)
        elif player_choice == "s":
            if gm_choice == player_choice:
                print(self.txt_draw)
            elif gm_choice == "r":
                self.gm_score += 1
                print(self.txt_gm_win)
            elif gm_choice == "p":
                self.player_score += 1
                print(self.txt_player_win)


    def print_instruction(self):
        print(self.instruction)


    def print_divider(self):
        print(self.divider)

    def clean_choice(self, text):
        new_text = text.replace("[", "").replace("]", "").lower()
        if len(new_text) > 1:
            if new_text == "rock":
                new_text = "r"
            elif new_text == "paper":
                new_text = "p"
            elif new_text == "scissor":
                new_text = "s"
            elif new_text != "exit" and new_text != "bye":
                new_text = ""
        return new_text


```

```python
Game().start()
```

# HW03 : ATM

```python
import random

class BankAccount:
    def __init__(self, name, pin, init_deposit: float):
        self.name = name
        self.pin = pin
        self.balance = init_deposit
        self.id = str(random.randint(10000, 99999))

    def deposit(self, amount):
        if amount <= 0:
            print("You cannot deposit lower than 0$.")
            return

        self.balance += amount
        print(f"{self.name}, you have deposited for {amount}$.")
        self.summary()

    def withdraw(self, amount):
        if amount <= 0:
            print("You cannot withdraw lower than 0$.")
            return
        elif amount > self.balance:
            print("You cannot withdraw more than your current balance.")
            return

        self.balance -= amount
        print(f"{self.name}, you have withdrawed for {amount}$.")
        self.summary()

    def summary(self):
        print(f"Your current balance is {self.balance}.")

```

```python
class ATM:
    divider = "------------------------------------"
    menus = [1,2,3,4,5]

    def __init__(self, bank_name):
        self.bank_name = bank_name
        self.accounts = []

    def start(self):
        print(f"Welcome to {self.bank_name}")
        is_exit = False

        while not is_exit:
            print("""Please choose menu:
[1] Open new account
[2] Withdraw
[3] Deposit
[4] Change Pin
[5] Close account
[e] Exit""")
            menu = input("").replace("[", "").replace("]", "").lower()
            if not menu.isdigit():
                is_exit = self.check_is_exit(menu)
                continue
            else:
                menu = int(menu)
                if not menu in self.menus:
                    print("Please try again.")
                    continue

                if menu == 1:
                    self.open_bank_account()
                    is_exit = True
                else:
                    account = self.login()
                    if account == None:
                        continue

                    if menu == 2:
                        self.withdraw(account)
                        is_exit = True
                    elif menu == 3:
                        self.deposit(account)
                        is_exit = True
                    elif menu == 4:
                        self.change_pin(account)
                        is_exit = True
                    elif menu == 5:
                        self.close_account(account)
                        is_exit = True

        print("Have a nice day.")




    def open_bank_account(self):
        user_name = input("What is your name? ")
        user_pin = ""
        user_init_deposit = 0
        is_exit = False
        while not is_exit:
            pin = input("Please define your 4-digit pin: ")
            if not pin.isdigit():
                is_exit = self.check_is_exit(pin)
                continue
            elif len(pin) != 4:
                continue
            user_pin = pin
            is_exit = True

        is_exit = False
        while not is_exit:
            deposit = input("How much of your first deposit? ")
            if not deposit.isdigit():
                is_exit = self.check_is_exit(deposit)
                continue

            user_init_deposit = float(deposit)
            is_exit = True

        user_account = BankAccount(user_name, user_pin, user_init_deposit)
        self.accounts.append(user_account)
        self.print_divider()
        print(f"""Your atm card no. is: {user_account.id}
Your balance: {user_account.balance}
Please use this atm card no. for other feature.
""")
        self.print_divider()
        return

    def login(self):
        card_id = input("Please insert your atm card no.: ")
        pin = input("Please insert your pin: ")

        for acct in self.accounts:
            if acct.id == card_id and acct.pin == pin:
                return acct

        print("Cannot find account. Please try again.")
        return None

    def withdraw(self, account: BankAccount):
        amount = input("How much do you want to withdraw? ")
        if not amount.replace(".", "").replace("-", "").isdigit():
            if self.check_is_exit(amount):
                return
            print("Please try again.")
            return
        else:
            self.print_divider()
            amount = float(amount)
            account.withdraw(amount)


    def deposit(self, account: BankAccount):
        amount = input("How much do you want to deposit? ")
        if not amount.replace(".", "").replace("-", "").isdigit():
            if self.check_is_exit(amount):
                return
            print("Please try again.")
            return
        else:
            self.print_divider()
            amount = float(amount)
            account.deposit(amount)


    def change_pin(self, account: BankAccount):
        is_exit = False
        while not is_exit:
            pin = input("Please define your new 4-digit pin: ")
            if not pin.isdigit():
                is_exit = self.check_is_exit(pin)
                continue
            elif len(pin) != 4:
                continue
            account.pin = pin

            is_exit = True
        print("Your pin has changed.")


    def close_account(self, account: BankAccount):
        print("This process cannot undo.")
        confirm_pin = input("If you want to continue, please insert your pin: ")
        if not confirm_pin.isdigit():
            if self.check_is_exit(confirm_pin):
                return
            print("Please try again.")
            return
        elif account.pin != confirm_pin:
            print("Incorrect pin. Cannot close account for now.")
            return
        else:
            self.accounts.remove(account)
            self.print_divider()
            print(f"{account.name} has closed account. Please receive your balance : {account.balance}")


    def check_is_exit(self, choice):
        print(f"choice : {choice}")
        return choice == "e" or choice == "exit" or choice == "cancel"

    def search(self, id):
        for acct in self.accounts:
            if acct.id == id:
                return acct
        return None

    def print_divider(self):
        print(self.divider)
```

```python
abc_atm = ATM("ABC Bank")
abc_atm.accounts.append(BankAccount("John", "0123", 45000))
abc_atm.accounts.append(BankAccount("Mary", "4567", 2000))
abc_atm.accounts.append(BankAccount("Alise", "8923", 67823.5))

for current_account in abc_atm.accounts:
    print(f"card no. {current_account.id} | name: {current_account.name}")
```

```python
abc_atm.start()
```
