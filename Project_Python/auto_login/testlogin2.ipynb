```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.maximize_window()
url = 'https://t.ly/3gJ4W'
driver.get(url)
time.sleep(3) # สร้าง delay 3 วิ
```

```python
df = pd.read_excel('data.xlsx')
df 
```

```python
for i in df.index:
    full_name_field = driver.find_element('xpath', '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[1]/div/div/div[2]/div/div[1]/div/div[1]/input')
    full_name_field.send_keys(df['Full name'][i]) # (df['Full name'][i]) คือการทำ loop df จาก code ด้านบน

    email_field = driver.find_element('xpath', '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[2]/div/div/div[2]/div/div[1]/div/div[1]/input')
    email_field.send_keys(df['Email'][i])

    mobile_number_field = driver.find_element('xpath', '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[3]/div/div/div[2]/div/div[1]/div/div[1]/input')
    mobile_number_field.send_keys('0'+str(df['Mobile number'][i])) #ใส่เลขตรงๆไม่ได้ ต้องแปลง type เป็น str

    university_field = driver.find_element('xpath', '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[4]/div/div/div[2]/div/div[1]/div/div[1]/input')
    university_field.send_keys(df['University/Organization'][i])

    submit_button = driver.find_element('xpath', '//*[@id="mG61Hd"]/div[2]/div/div[3]/div[1]/div[1]/div/span/span')
    submit_button.click()
    driver.back()
```
